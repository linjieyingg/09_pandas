```python
df_data = pd.read_excel('airlines.xlsx')
df_data
```

```python
df_data.dropna(how='any', inplace=True)
df_data
```

```python
df1 = df_data[df_data.isna().any(axis=1)]
df1
```

```python
df_data['Date_of_Journey'] = pd.to_datetime(df_data['Date_of_Journey'], errors='coerce')
```

```python
df_data['Dep_Time'] = pd.to_datetime(df_data['Dep_Time'], format='%H:%M').dt.time
```

```python
df_data['Arrival_Time'] = pd.to_datetime(df_data['Arrival_Time'], format='mixed').dt.time
df_data
```

```python
df_data['Journey_day'] = df_data['Date_of_Journey'].dt.day_name()
df_data['Journey_month'] = df_data['Date_of_Journey'].dt.month_name()
df_data['Journey_year'] = df_data['Date_of_Journey'].dt.year
df_data['Dep_Time_hour'] = pd.to_datetime(df_data['Dep_Time'].astype(str)).dt.hour
df_data['Dep_Time_minute'] = pd.to_datetime(df_data['Dep_Time'].astype(str)).dt.minute
df_data['Arrival_Time_hour'] = pd.to_datetime(df_data['Arrival_Time'].astype(str)).dt.hour
df_data['Arrival_Time_minute'] = pd.to_datetime(df_data['Arrival_Time'].astype(str)).dt.minute
df_data
```

```python
def check(hour):
    if (hour >= 0 and hour<4 ):
        return 'Late Night'
    elif (hour < 8):
        return 'Early Morning'
    elif (hour < 12):
        return 'Morning'
    elif (hour < 16):
        return 'Noon'
    elif (hour < 20):
        return 'Evening'
    elif (hour < 24):
        return 'Night' 
```

```python
df_data['dep_description'] = df_data.apply(lambda row: check(row['Dep_Time_hour']),axis=1)
```

```python
df_data
```

```python
data = df_data['dep_description'].value_counts()
data.plot.bar(ylabel='Number of flights')
```

```python
def total(hours, minutes):
    try:
        hours = int(hours)
    except:
        hours = 0
    try:
        minutes = int(minutes)
    except:
        minutes = 0
    return hours * 60 + minutes
```

```python
df_data['Duration_hours'] = df_data['Duration'].str.split('h').str[0]
df_data['Duration_mins'] = df_data['Duration'].str.split('h').str[1].str.split('m').str[0]
df_data['Duration_mins'][df_data['Duration_mins']== ''] = 0
df_data['Duration_total_mins'] = df_data.apply(lambda row: total(row['Duration_hours'], row['Duration_mins']),axis=1)
df_data
```

```python
m,b = np.polyfit(df_data['Duration_total_mins'],df_data['Price'],1)
```

```python
import matplotlib.pyplot as plt
```

```python
data = df_data.plot.scatter(x='Duration_total_mins', y='Price', colormap='jet')
plt.axline(xy1=(0, b), slope=m, label=f'$y = {m:.1f}x {b:+.1f}$')
```

Longer times between departure and arrival usually means higher prices for the flight(s), however there are many outliers.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
from patsy import dmatrices
```

```python
df_price_min = dmatrices("Price ~ Duration_total_mins - 1", df_data, return_type='dataframe')
```

```python
df_price_min
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    df_price_min[1], df_price_min[0], test_size=0.20, random_state=42)
```

```python
len(x_train), len(y_train), len(x_test), len(y_test)
```

```python
model_price_min = LinearRegression()
model_price_min.fit(x_train, y_train)
```

```python
plt.scatter(x_train['Duration_total_mins'], y_train, label='Price Duration_total_min')
x_min, x_max = plt.xlim()
plt.plot([x_min, x_max],
         [model_price_min.coef_[0][0]*x_min+ model_price_min.intercept_[0],
          model_price_min.coef_[0][0]*x_max+ model_price_min.intercept_[0]],
          label="Linear regression", color="red")
```
